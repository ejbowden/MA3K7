```python
#importing libraries
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#creating function to simulate process
def shorten(length):
    x=list(range(1,length+1))
    #print(x)
    while len(x)>1:
        index1=random.randint(0,len(x)-1)
        z1=x[index1]
        del x[index1]
        index2=random.randint(0,len(x)-1)
        z2=x[index2]
        del x[index2]
        z3=np.abs(z2-z1)
        x.append(z3)
        #print(x)
    return x[0]
```

```python
for i in range(10):
    x=shorten(2024)
    print(x)
```

```python
#creating graph of possibilities for n=2024
k=2024
l=np.linspace(1,k,k+1)
m=np.zeros(k+1)

for i in range(10000):
    product=shorten(k)
    m[product]+=1
    
plt.bar(l,m)
plt.ylabel('Times obtained')
plt.xlabel('Result')
plt.show()
```
