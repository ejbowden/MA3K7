```python
#function for Beth's turn - entirely computer decided
def bethTurn(glasses):
    highest=0.0
    empty1=1
    l=len(glasses)
    #finding fullest glass
    for i in range(l):
        if glasses[i]>highest:
            highest=glasses[i]
            empty1=i
    #finding fullest adjacent glass
    if empty1==0:
        empty2=l-1
        if glasses[(empty1+1)]>glasses[empty2]:
            empty2=empty1+1
    elif empty1==l-1:
        empty2=0
        if glasses[(empty1-1)]>glasses[empty2]:
            empty2=empty1-1
    else:
        empty2=empty1-1
        if glasses[(empty2+2)]>glasses[empty2]:
            empty2=empty2+2
    #emptying glasses
    glasses[empty1]=0.0
    glasses[empty2]=0.0
```

```python
#function for Ali's turn - requires user input
def aliTurn(glasses):
    l=len(glasses)
    total=0
    #adding water to each glass
    for i in range(l):
        validAmount=False
        while validAmount==False:
            x=float(input('How much water do you want to add to this glass?'))
            if total+x<=0.5:
                validAmount=True
            else:
                print("Not a valid option - please try again.")
        glasses[i]+=x
        total+=x
```

```python
bethWin=False
aliWin=False
#setting number of glasses
n=6
#creating list of glasses
glasses=[]
for i in range(n):
    glasses.append(0.0)
print(glasses)

while bethWin==False and aliWin==False:
    #ali's turn
    aliTurn(glasses)
    print(glasses)
    l=len(glasses)
    for i in range(l):
        if glasses[i]>1:
            aliWin=True
    #beth's turn if ali hasn't won
    if aliWin==False:
        totalEmpty=0
        bethTurn(glasses)
        print(glasses)
        for i in range(l):
            if glasses[i]==0:
                totalEmpty+=1
        if totalEmpty==l:
            bethWin=True

if aliWin==True:
    print("Ali wins!")
if bethWin==True:
    print("Beth wins!")
```
