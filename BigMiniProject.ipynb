```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Strategy 1 (partially filling all glasses at the start)
def strat1(noGlasses):
    n=[]
    k=0
    while k<((noGlasses-3)/2)+1:
        n.append(k)
        k+=1
    total=0
    for i in n:
        total+=(1/(noGlasses-2*n[i]))
    total=total/2
    return total
```

```python
x=3
list1=[]
list2=[]
while x<50:
    list1.append(x)
    y=strat1(x)
    list2.append(y)
    x+=2
```

```python
for i in range(len(list1)):
    if list2[i]>1/2:
        print(2*i+3)
        break
```

```python
y=strat1(13)
print(y)
y=strat1(15)
print(y)
```

```python
plt.plot(list1,list2)
y=list1.copy()
for i in range(len(list1)):
    y[i]=0.5
plt.plot(list1,y)
plt.xlabel('Number of glasses')
plt.ylabel('Amount of water in each glass')
plt.show()
```

```python
#Strategy 2 - (filling alternate glasses at the start)
def strat2(noGlasses):
    n=[]
    k=0
    while k<((noGlasses/2)+1):
        n.append(k)
        k+=1
    total=0
    for i in range(len(n)-2):
        total+=(1/(n[i+2]*2))
    return total
strat2(8)
```

```python
x=2
list1=[]
list2=[]
while x<40:
    list1.append(x)
    y=strat2(x)
    list2.append(y)
    x+=2
```

```python
for i in range(len(list1)):
    if list2[i]>1/2:
        print(list1[i])
        break
```

```python
plt.plot(list1,list2)
y=list1.copy()
for i in range(len(list1)):
    y[i]=0.5
plt.plot(list1,y)
plt.xlabel('Number of glasses')
plt.ylabel('Amount of water in each glass')
plt.show()
```

```python
#strategy 4 - code for w*sum 1 to n of 1/k - plot w against n for sum to be 1
def strat4(noGlasses):
    n=int(noGlasses/2)
    total=0
    for k in range(n):
        x=1/(k+1)
        total+=x
    w=1/total
    return w

list1=[]
n=100
k=2
while k<n:
    list1.append(k)
    k+=2
        
list2=list1.copy()
for i in range(0,len(list2)):
    list2[i]=strat4(list2[i])

plt.plot(list1,list2)
plt.xlabel('Number of glasses used')
plt.ylabel('Volume of water added')
plt.ylim(0,1.1)
plt.show()
```
