```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#function to simulate the process happening once up to a certain number
def model(aim):
    position=1
    while position<aim:
        x=random.randint(1,2)
        position+=x
    if position==aim:
        return 1
    else:
        return 0
```

```python
#function to repeat the process and count how many times it lands on the chosen number
def repeat(aim):
    total=0
    for i in range(100000):
        x=model(aim)
        if x==1:
            total+=1
    return total
```

```python
#running the process for different aims and plotting a graph of how often they occur
k=10
x=np.linspace(1,k,k)
y=x.copy()
for i in range(len(x)):
    y[i]=repeat(x[i])/100000
z=x.copy()
for i in range(k):
    z[i]=2/3

plt.plot(x,z)
plt.plot(x,y)
plt.xlabel("Aim")
plt.ylabel("Probability of landing on aim")
plt.show()
```

```python
def modelDie(aim,sides):
    position=1
    while position<aim:
        x=random.randint(1,sides)
        position+=x
    if position==aim:
        return 1
    else:
        return 0

def repeatDie(aim,sides):
    total=0
    for i in range(100000):
        x=modelDie(aim,sides)
        if x==1:
            total+=1
    return total

k=100
x=np.linspace(1,k,k)
w=x.copy()
y=x.copy()
z=x.copy()
v=x.copy()
for i in range(len(x)):
    w[i]=repeatDie(x[i],2)/100000
    y[i]=repeatDie(x[i],6)/100000
    z[i]=repeatDie(x[i],12)/100000
    v[i]=repeatDie(x[i],20)/100000
    
plt.plot(x,w, label="2")
plt.plot(x,y, label="6")
plt.plot(x,z, label="12")
plt.plot(x,v, label="20")
plt.legend(fontsize=10)
plt.xlabel("Aim")
plt.ylabel("Probability of landing on aim")
plt.show()
```
